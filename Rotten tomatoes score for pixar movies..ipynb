```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].str.replace('%','')
df
```

```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].astype('int64')
df.dtypes
```

```python
plt.figure(figsize=(20,10))

x=df['Moviename']
y=df['Rotten Tomatoes']

plt.bar(x,y,color='#F47340')

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center',va='bottom',fontname='Times New Roman',fontsize=17)
        
addlabels(x,y)

plt.title("Rotten Tomatoes rating for Pixar Movies",fontname='Times New Roman',fontsize=20,pad=15)
plt.xlabel("Movies",fontname='Times New Roman',fontsize=20,labelpad=15,color='red')
plt.ylabel("Percentage",fontname='Times New Roman',fontsize=20,labelpad=15,color='red')


plt.xticks(fontname='Times New Roman',fontsize=18,rotation=90)
plt.yticks(fontname='Times New Roman',fontsize=18)

plt.show()
```
